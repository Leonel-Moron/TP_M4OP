```python
filtro = (dataframe["distrito_nombre"] == "Ciudad Autónoma de Buenos Aires") & (dataframe["cargo_nombre"] == "PRESIDENTE Y VICE") & (dataframe["agrupacion_nombre"] == "UNION POR LA PATRIA") & (dataframe["seccion_nombre"] == "Comuna 01")
resultado = dataframe.loc[filtro, "votos_cantidad"].sum()
resultado
#Son los votos de la agrupacion en CABA, aparece despues de filtrar. la idea ahora es incluir ese resultado en la tabla


```

```python
dataframe.groupby(["agrupacion_nombre","votos_cantidad"]).size().to_frame().head(5)
#Agrupamos las agrupaciones y nos indica los votos que tuvo cada uno (pero no es la suma)
#va a mostrar que ACCION VILLARINO tuvo una fila con 29 votos 1 sola vez, o una fila de 39 votos 2 veces, NO SON LOS VOTOS TOTAL
```

```python
dataframe.groupby(["agrupacion_nombre"])["votos_cantidad"].sum().to_frame()
#Ahora si es la suma (se puede sacar promedio, producto, etc, etc)
#Primero se agrupa por agrupacion y despues se calcula la suma de los votos_cantidad
#Si se agrega otro argumento dentro del groupBy, va a cambiar el dataframe
```

```python
dataframe.groupby(["agrupacion_nombre","mesa_tipo"])["votos_cantidad"].mean().to_frame()
```

```python
dataframe.groupby(["agrupacion_nombre"])["votos_cantidad"].max().to_frame()
#Va a devolover un df mostrando la mayor cantidad (el valor maximo) de votos que hubo en una fila
```
