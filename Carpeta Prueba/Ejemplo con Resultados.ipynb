```python
dataframe.groupby(["agrupacion_nombre","votos_cantidad"]).size().to_frame().head(5)
#Agrupamos las agrupaciones y nos indica los votos que tuvo cada uno (pero no es la suma)
#va a mostrar que ACCION VILLARINO tuvo una fila con 29 votos 1 sola vez, o una fila de 39 votos 2 veces, NO SON LOS VOTOS TOTAL
```

```python
dataframe.groupby(["agrupacion_nombre"])["votos_cantidad"].sum().to_frame()
#Ahora si es la suma (se puede sacar promedio, producto, etc, etc)
#Primero se agrupa por agrupacion y despues se calcula la suma de los votos_cantidad
#Si se agrega otro argumento dentro del groupBy, va a cambiar el dataframe
```

```python
dataframe.groupby(["agrupacion_nombre","mesa_tipo"])["votos_cantidad"].mean().to_frame()
```

```python
dataframe.groupby(["agrupacion_nombre"])["votos_cantidad"].max().to_frame()
#Va a devolover un df mostrando la mayor cantidad (el valor maximo) de votos que hubo en una fila
```

```python
#Con agrregate (agg) podemos incluir mas medidas de agregacion
#Se agrupa x las diferentes agrupaciones, y se aplica a votos cantidad las operaciones
dataframe.groupby("agrupacion_nombre")["votos_cantidad"].agg(["max","min","sum","std"])
```

```python
dataframe.groupby("agrupacion_nombre")["votos_cantidad"].agg(maximo ="max",minimo ="min", variacion_std = "std", total_votos ="sum")
#Lo mismo que arriba pero mostramos que podemos asignar nombres a los resultados en vez de usar los x default
#Puedo hacer calculos de diferentes columnas a la vez tmb (pero en ese df solo hay votos para trabajar integers)
```

```python
#Vamos a filtrar segun algunas variables. En este caso por votos, solo vamos a mostrar las agrupaciones que pasen un umbral de votos >60000
filtro = dataframe.groupby("agrupacion_nombre")["votos_cantidad"].agg(votos_total = sum)
#Primero agrupamos por agrupaciones y sumamos su total de votos, ademas le ponemos nombre a todo se va a ver la suma
```

```python
filtro
```

```python
filtro.columns
#Vamos a ver que hay solo una columna (votos_total), agrupacion_nombre ahora es el Index
#Hay que tratar que sea una columna las agrupaciones y los index sean numerados como antes
```

```python
#creamos otro df usando el df filtrado. Indicamos que muestre los mayores a 60k.. y ademas le reseteamos el index para que las agrupaciones 
#sean columna en este nuevo df
filtro2= filtro.loc[(filtro["votos_total"] > 60000)].reset_index()
```

```python
filtro2.shape
#Corroboramos que ahora hay 2 COLUMNAS y no solo 1 como antes, y ademas hay menos filas, o sea, funciono.
```
