```python
#VAMOS A HACER UN MERGE DE 2 DF

kcal= {"Marca":["Coca", "Coca Zero", "Pepsi", "Manaos"],
       "Kcal":[37, 0.1, 31, 45]
    }

kcal= pd.DataFrame(kcal).sort_values(by="Marca").reset_index(drop=True)
kcal
```

```python
macro= {"Marca":["Coca", "Pepsi", "Cunnington", "Manaos", "Sprite"],
        "Azucares":[9, 7.5, 10, 11,10],
        "Proteinas": [0.1, 0, 0.2, 0.1,0],
        "Grasas": [0, 0, 0, 0,0] 
    }

macro= pd.DataFrame(macro).sort_values(by="Marca").reset_index(drop=True)
macro
```

```python
gaseosas_inner = pd.merge(kcal, macro, on="Marca", how="inner")
gaseosas_inner
#Se usa la funcion merge() y se pasa por parametro los df a mergear, sobre que unirlos, y que tipo de union
#Con el inner se muestran solo los que coinciden en las filas. Tiene que estar en las 2 filas de los DF el valor para un inner
#No va a tener NaN porque coinciden las filas, ninguna prevalece sobre la otra.
```

```python
gaseosas_left = pd.merge(kcal, macro, on="Marca", how="left")
gaseosas_left
#Con el left se copia el df izquierdo y solo se copian las coincidencias del df de la derecha, sino se desechan
#La copia se hace sobre las filas coincidentes. Con ello trae las columnas, y rellena con NaN si no tiene datos en esas columnas
#Coca zero en el DF left solo tiene el valor de KCAL, el resto de las columnas es del DF right (entonces las rellena con NaN)
```

```python
gaseosas_right = pd.merge(kcal, macro, on="Marca", how="right")
gaseosas_right
#Con el right es lo mismo, se copia el derecho, y si coinciden las filas del DF izquierda.
#Los valores que no tenga el izquierdo se llena con NaN
```

```python
gaseosas_outer = pd.merge(kcal, macro, on="Marca", how="outer")
gaseosas_outer
#Outer directamente une los 2 DF, no importa las coincidencias
```

```python
finanzas = {"Nombre": ["Juan Pérez", "María Gómez","Juan Pérez","Leonel Morón"], "Area": ["Finanzas", "Finanzas","Finanzas","At cliente"]}
finanzas = pd.DataFrame.from_dict(finanzas)
rrhh = {"Nombre": ["Ana Estévez", "Diego Hernández","Marcela Pérez García","Leonel Morón"], "Area": ["RRHH", "RRHH","RRHH", "At cliente"]}
rrhh = pd.DataFrame.from_dict(rrhh)
ventas= {"Nombre": ["Matías Rodríguez", "Silvia García"], "Area": ["Ventas", "Ventas"],"Antiguedad_Anios": [3, 6.9]}
ventas= pd.DataFrame.from_dict(ventas)
#Creamos 3 DF

empleados_union_all = pd.concat([finanzas, rrhh, ventas],ignore_index=True)
#Usamos la funcion (parecida a merge pero te une todo).
#Ignore_index ayuda a que se eliminen los indices viejos y ahora tenga nuevos.
empleados_union_all 

#UNION ALL ignora los duplicados, o sea, no les da bola y copia tal cual.
#Se van a unir 1 por debajo del otro los DF, es lo mas simple. Si hay columnas coincidentes se llenan los datos, sino NaN
```

```python
empleados_union = pd.concat([finanzas, rrhh, ventas]).drop_duplicates().reset_index(drop=True)
#agregar la funcion drop_duplicates hace que los valores duplicados se eliminen, asi se limpia el DF nuevo
#Se usa los mismos DF para crear un nuevo pero van a aparecer menas filas
#UNION elimina los registros duplicados
empleados_union
```
