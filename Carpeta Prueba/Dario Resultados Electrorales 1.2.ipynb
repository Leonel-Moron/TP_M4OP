```python
'''EL DATASET VIENE CON DISTINTOS VALORES QUE NOS PUEDE GENERAR ROMPER EN ALGUN CODIGO, 
POR LO QUE VAMOS A PASARLOS A 'STRING' PARA PODER MANIPULARLOS MEJOR'''

todo = Resultados [["distrito_id","distrito_nombre","seccion_id","seccion_nombre","circuito_id","mesa_id","mesa_tipo","mesa_electores","cargo_id","cargo_nombre","agrupacion_nombre","votos_tipo","votos_cantidad"]]
todo[["distrito_nombre", "seccion_nombre","circuito_id","mesa_tipo","cargo_nombre","agrupacion_nombre","votos_tipo"]] = todo[["distrito_nombre", "seccion_nombre","circuito_id","mesa_tipo","cargo_nombre","agrupacion_nombre","votos_tipo"]].astype("string")
```

```python
'''TENEMOS UNA COLUMNA QUE ES 'VOTOS TIPO' Y OTRA 'AGRUPACION'. CORREGIMOS EL HECHO DE QUE CUANDO SE NOS APARECE UN VOTO
RECURRIDO EN VOTO TIPO Y LUEGO EN AGRUPACIÓN APARECE COMO 'NA'.LA FUNCION BUSCA QUE CADA VEZ QUE HAYA UN VALOR 'NA' 
ESTA SEA REEMPLAZADA POR EL VALOR QUE VIENE EN 'TIPO DE VOTO'.'''

def replace_nan_with_corresponding_value(df):
    df['agrupacion_nombre'].fillna(df['votos_tipo'], inplace=True)
    return df
replace_nan_with_corresponding_value(todo)
```

```python
'''AQUÍ ESTAMOS TOMANDO CABA SOLAMENTE DEL DATAFRAME, Y SOLO LAS COLUMNAS DE LA 2 A LA 13'''

CABA =(todo[todo["distrito_id"]==1])
CABA = CABA[CABA.columns[2:13]]
```

```python
'''TENEMOS QUE DIFERENCIAR LOS VOTOS QUE NECESITAMOS, POR LO QUE ESPECIFICAMOS EL VOTO
DE PRESIDENTE Y VICE SOLO PARA CABA, ADEMÁS DE TOMAR SOLO LOS VOTOS DE UNION POR LA PATRIA
'''

(
    CABA[CABA["agrupacion_nombre"]=="UNION POR LA PATRIA"]
    [CABA["cargo_nombre"]=="PRESIDENTE Y VICE"]
    .groupby(["agrupacion_nombre","seccion_nombre"])[["votos_cantidad"]].sum()
    .sort_values(by="votos_cantidad")

)
```

```python
''' Tomamos los votos de UP y los transformamos en porcentajes para cada comuna ''' 

porcentaje_x_comunaUP = (
    CABA[["seccion_nombre","votos_cantidad"]]
    .groupby(["seccion_nombre"])
    .sum()
    .transform(lambda x: (x + 0.0) / x.sum() * 100)
)
print(porcentaje_x_comunaUP)
```

```python
dpto = dpto[["DPTO_REF_ID", "NOMDPTO"]]
dpto
```

```python
dpto_frac = pd.merge(dpto, frac, on="DPTO_REF_ID", how="left")
dpto_frac.groupby = ["NOMDPTO"]
dpto_frac
```

```python
dpto_frac_radio = pd.merge(dpto_frac, radio, on="FRAC_REF_ID", how="right")
dpto_frac_radio
```

```python
dpto_frac_radio_vivienda = pd.merge(dpto_frac_radio, vivienda, on="RADIO_REF_ID", how="right")
dpto_frac_radio_vivienda
```

```python
CONJUNTOhogar =  pd.merge(dpto_frac_radio_vivienda, hogar, on="VIVIENDA_REF_ID", how="right")
CONJUNTOhogar
```

```python
CONJUNTOpersonas = pd.merge(CONJUNTOhogar, persona, on="HOGAR_REF_ID", how="right")
votantes = CONJUNTOpersonas[CONJUNTOpersonas["P03"] >= 18]
votantes
```
